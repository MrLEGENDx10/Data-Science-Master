```python
import numpy as np
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
```

```python
plt.scatter(x,y)
```

```python
plt.scatter(x,y)
plt.xlabel("This is x axis")
plt.ylabel("This is y axis")
```

```python
plt.scatter(x,y , c= '#FF0560')
plt.xlabel("This is x axis")
plt.ylabel("This is y axis")
plt.title("This is a title") 
```

```python
plt.scatter(x,y , c= '#FF0560', alpha=0.5)
plt.xlabel("This is x axis")
plt.ylabel("This is y axis")
plt.title("This is a title") 
```

```python
plt.scatter(x,y , c= '#FF0560',alpha=0.5, marker='^')
plt.xlabel("This is x axis")
plt.ylabel("This is y axis")
plt.title("This is a title") 
```
