# Feature Scaling: A Critical Step in Machine Learning

**Feature Scaling** is a critical step in building accurate and effective machine learning models. One key aspect of feature engineering is **scaling**, **normalization**, and **standardization**. These techniques involve transforming the data to make it more suitable for modeling. 

These techniques can help to:

- Improve model performance
- Reduce the impact of outliers
- Ensure that the data is on the same scale

In this article, we will explore the concepts of scaling, normalization, and standardization. We'll discuss why they are important and how to apply them to different types of data. 

By the end of this article, you'll have a thorough understanding of these essential feature engineering techniques. You'll be able to apply them to your own machine learning projects.

```python
import seaborn as sns
```

```python
df= sns.load_dataset('tips')
```

```python
df.head()
```

```python
import numpy as np
mean = np.mean(df['total_bill'])
std = np.std(df['total_bill'])
print(mean,std)
```

```python
normalized_data = []
for i in list(df['total_bill']):
    z_score = (i-mean)/std
    normalized_data.append(z_score)
```

```python
print(normalized_data)
```

```python
sns.histplot(df['total_bill'])
```

```python
sns.histplot(normalized_data)
```

The line of code you've provided is using the `fit` method of a scaler object from the `scikit-learn` library on the `total_bill` column of the DataFrame `df`.

Here's a breakdown of what this code is doing:

- `df[['total_bill']]`: This is selecting the `total_bill` column from the DataFrame `df`. The double brackets (`[[ ]]`) mean that the output will be another DataFrame.

- `scaler.fit()`: This is calling the `fit` method of the `scaler` object. The `fit` method is used to compute the minimum and maximum values to be used for later scaling. Basically, it's calculating the parameters it needs to perform the actual scaling.

The `scaler` object is likely an instance of one of the scaler classes from `scikit-learn`, such as `MinMaxScaler`, `StandardScaler`, or `RobustScaler`. These are used to transform the data to fit certain statistical properties, such as a specific range or a zero mean and unit variance.

The actual scaling (transformation) of the data would be done with the `transform` method, often in the form `scaler.transform(df[['total_bill']])`.

The provided code is performing feature scaling on the 'total_bill' and 'tip' columns of a DataFrame `df` using a scaler object from the `scikit-learn` library. Here's a breakdown:

- `df[['total_bill','tip']]`: This is selecting the 'total_bill' and 'tip' columns from the DataFrame `df`. The output will be another DataFrame.

- `scaler.fit_transform()`: This method first fits the scaler to the data by computing the necessary transformation parameters (like mean and standard deviation for `StandardScaler`, or min and max values for `MinMaxScaler`), and then it transforms the data using these parameters. The transformed data will have the property that the scaler enforces, like zero mean and unit variance for `StandardScaler`, or values between 0 and 1 for `MinMaxScaler`.

- `pd.DataFrame()`: This is converting the output of `scaler.fit_transform()` back into a DataFrame. The `columns=['total_bill','tip']` argument is used to name the columns of the new DataFrame.
