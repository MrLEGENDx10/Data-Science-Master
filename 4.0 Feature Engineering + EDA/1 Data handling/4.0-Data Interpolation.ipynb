## Data Interpolation
Data interpolation is the process of estimating unknown values within a dataset based on the known values. In Python, there are various libraries available that can be used for data interpolation, such as NumPy, SciPy, and Pandas. Here is an example of how to perform data interpolation using the NumPy library:

1. Linear Interpolation
2. Cubic Interpolation With Scipy
3. Polynomial Interpolation

## 1. Linear Interpolation

```python
import numpy as np
x = np.array([1, 2, 3, 4, 5])
y= np.array([2,4,6,8,10])
```

```python
import matplotlib.pyplot as plt
plt.scatter(x, y)
```

```python
x_new = np.linspace(1, 5, 100)
# The line `y_interp = np.interp(x_new,x,y)` is using linear interpolation to estimate the values of `y` at the points specified by `x_new`.
y_interp = np.interp(x_new,x,y)
```

```python
y_interp
```

```python
figure, ax = plt.subplots(figsize=(50,20))
ax.scatter(x_new, y_interp)
plt.show()
```

## 2. Cubic Interpolation With Scipy

```python
import numpy as np
x = np.array([1, 2, 3, 4, 5])
y= np.array([1,8,27,64,125])
```

```python
from scipy.interpolate import interp1d
```

```python
f = interp1d(x,y,kind='cubic')
```

```python
x_new = np.linspace(1, 5, 10)
y_interp = f(x_new)
```

```python
y_interp
```

```python
plt.scatter(x_new, y_interp)
```

## 3. Polynomial Interpolation

```python
import numpy as np

x = np.array([1, 2, 3, 4, 5])
y = np.array([1,4,9,16,25])
```

```python
p= np.polyfit(x,y,2)
```

```python
x_new = np.linspace(1, 5, 10)
y_interp = np.polyval(p,x_new)
```

```python
plt.scatter(x_new, y_interp)
```
