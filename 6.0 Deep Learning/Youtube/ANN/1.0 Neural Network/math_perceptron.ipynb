```python
import numpy as np 
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')

x_axis = np.linspace(-2,2,200)
# The line `step = np.where(x_axis<0,1,0)` is creating an array `step` where the value is 1 if the corresponding element in `x_axis` is less than 0, and 0 otherwise.
step = np.where(x_axis<0,0,1)

plt.plot(x_axis,step)
plt.xlabel("X-axis")
plt.ylabel(r"$step(z)$")
plt.title("Step Function")
plt.axhline(0,color='k',lw= 1)
plt.axvline(0,color='k',lw= 1)
```

* $$sgn(z) = \left\{\begin{matrix}
  -1 & z < 0 \\
  0 & z = 0 \\
  1 & z > 0                                
  \end{matrix}\right.$$

```python
# This code snippet defines a function `sgn(x)` that returns the sign of a number `x`. If `x` is less than 0, it returns -1; if `x` is greater than 0, it returns 1; and if `x` is equal to 0, it returns 0.
def sgn(x):
    if x<0:
        return -1
    elif x>0:
        return 1
    return 0

# This line of code is applying the `sgn` function to each element in the `x_axis` array, and then converting the resulting list of sign values into a NumPy array using `np.array()`.
sgn=np.array(list(map(sgn,x_axis)))

plt.plot(x_axis,sgn)
plt.xlabel("X-axis")
plt.ylabel(r"$sgn(z)$")
plt.title("Signum Function")
plt.axhline(0,color='k',lw= 1)
plt.axvline(0,color='k',lw= 1)

```

* These activation function are **analogous to firing of neurotransmitter** at a specific threshold
* Perceptron learning Rule:
$$w_{i,j} \leftarrow w_{i,j} + \eta(y_j - \hat{y_j})x_i$$
where,
  * $w_{i,j}$ : connection weight between $i^{th}$ input neuron and $j^{th}$ output neuron
  * $x_i$ : $i^{th}$ input value.
  * $\hat{y_j}$ : output of $j^{th}$ output neuron
  * $y_j$ : target output of $j^{th}$ output neuron
  * $\eta$ : learning rate
    
* It can also be written as for jth element of w vector
    $$w_j = w_j + \triangle w_j$$
    $$where,\ \triangle w_j =  \eta(y^{(i)} - \hat{y_j}^{(i)})x_j^{(i)}$$


* Single TLUs (Threshold Logic Unit) are **simple linear binary classifier** hence not suitable for non linear operation.
* Rosenblatt proved that if the data is **linearly separable** then only this algorithm will converge which is known as **Perceptron learning theorem**

## Derivation :-

Let's assume that you are doing a binary classification with class +1 and -1

Let there be decision function $\phi (z)$

which takes linear combination of certain inputs "$\mathbf{x}$"

corresponding weights "$\mathbf{w}$" and net input $\mathbf{z} = w_1 x_1 + w_2 x_2 + ... + w_n x_n$


So in vector form we have

$\begin{matrix}
\mathbf{w} = \begin{bmatrix}
w_1\\
\vdots\\
w_n\end{bmatrix}&&
\mathbf{x} = \begin{bmatrix}
x_1\\
\vdots\\
x_n\end{bmatrix}\end{matrix}$


so,  $\mathbf{z = w^T x}$

Now, if

for a sample $\mathbf{x}$

$\phi(z) = \begin{cases}
+1 & \text{ if } z \geq \theta \\
-1 & \text{ if } z< \theta
\end{cases}$

Lets simplify the above equation -

$\phi(z) = \begin{cases}
+1 & \text{ if } z - \theta \geq 0\\
-1 & \text{ if } z - \theta < 0
\end{cases}$

Suppose $w_0 = - \theta$ and $x_0 = 1$

Then,

$\mathbf{z'} = w_0 x_0 + w_1 x_1 + w_2 x_2 + ... + w_n x_n$

and

$\phi(z) = \begin{cases}
+1 & \text{ if } z \geq 0\\
-1 & \text{ if } z < 0
\end{cases}$


```python
### Implementing of perceptron with python
import numpy as np
import matplotlib.pyplot as plt
import joblib as jb 
# `joblib` can be used to save trained models to disk so that they can be easily reloaded and used for predictions without having to retrain the model each time.
import pandas as pd
plt.style.use('fivethirtyeight')
```

```python
class Perceptron:
    def __init__(self,eta,epochs):
        self.weights =np.random.randn(3) * 1e-4
        # `np.random.randn(3)` generates an array of 3 random numbers from a standard normal distribution (mean=0, variance=1). 
        self.eta=eta
        self.epochs=epochs
        
    def activationFunction(self,input,weights):
        z = np.dot(input,weights)
        return np.where(z>0,1,0)
    
    def fit(self,X,y):
        self.X=X
        self.y=y
        print(f'Original X: \n{X}')
        print(f'Original y: \n{y}')
        X_with_bias = np.c_[self.X, -np.ones((len(self.X), 1))]
        print(f'X with bias: \n{X_with_bias}')
        
        for epoch in range(self.epochs):
            print(f"for epoch: {epoch}")
            y_hat = self.activationFunction(X_with_bias,self.weights)
            print(f'predicted value after forward pass: \n{y_hat}')
            error = self.y - y_hat
            print(f'Error: \n{error}') 
            print(f"Previous weights: {self.weights}") 
            self.weights = self.weights + self.eta * np.dot(X_with_bias.T,error)
            print(f'updated weights: \n{self.weights}')
            print("####"*10)
            
    def predict(self,X):
        X_with_bias = np.c_[X,-np.ones((len(self.X),1))]
        return self.activationFunction(X_with_bias,self.weights)
```

```python
## AND GATE

data = {
    'x1': [0,0,1,1],
    'x2': [0,1,0,1],
    'y': [0,0,0,1]
}
AND=pd.DataFrame(data)
AND
```

```python
X = AND.drop('y',axis=1)
y = AND['y']
y.to_frame()
```

```python
model = Perceptron(eta=0.01,epochs=10)
model.fit(X,y)
```

```python
print(model.predict(X))
```
